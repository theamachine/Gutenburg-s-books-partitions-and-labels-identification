Main function "book" using regular expression operation to clean the text, make 200 partitions for the book each of 100 words, returning pandas dataframe that has "partitions" and their "labels".

```python
def book (book_name):
    book_row = gutenberg.raw(book_name)
    regex_output = re.sub('[^a-z0-9\']+', ' ', book_row.lower())
    words = regex_output.split()
    partition  = 0
    lst  = []
    word_index = 0
    line = 0
    while partition < 200:
        word_count=0
        txt = []
        txt = " ".join(words[partition*100:partition*100+100])
        partition +=1
        lst.append(txt) 
    pandas = pd.DataFrame({ "partition": lst, "Label": [book_name[:-4]]*200})
    return pandas
```
